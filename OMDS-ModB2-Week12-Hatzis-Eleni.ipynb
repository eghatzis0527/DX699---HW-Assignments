# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

For chart 1, one advantage is you can compare them side by side and not have to see them overlapping. However, there are couple major disadvantages of this graph. To start, the first disadvantage is the y-axes are not the same and have different units. This makes the graphs look like they are on the same scale and can mislead the audience. Furthermore, the labels are only on the right most graph and are going into the left graph which can be distracting for the audience.

For chart 2, there are a lot of advantages to this graph. The two lines being on the same axis properly shows the differences between the lines so there isn't any confusion on the axes not being the same as they were in the first graph. Furthermore, the labels are not going into the graph so it isn't as distracting for the audience. The only disadvantage is there is overlapping and no label in the chart, so the audience isn't able to tell which line is which. This means you can't tell which line corresponds to which week and whether there was an increase or decrease in vistors in the next week. 

For both charts, we would need to be sure to add titles to them for the audience and set the y-axes to 0 so they are not creating any confusion for the audience.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python
import pandas as pd

prod_a = {'2008' : 400, '2009' : 410, '2010' : 420, '2011' : 375, '2012' : 250, '2013' : 225, '2014' : 210}
prod_b = {'2008' : 350, '2009' : 390, '2010' : 400, '2011' : 365, '2012' : 245, '2013' : 220, '2014' : 205}
prod_c = {'2010' : 100, '2011' : 150, '2012' : 190, '2013' : 210, '2014' : 190}
prod_d = {'2011' : 145, '2012' : 250, '2013' : 210, '2014' : 200}
prod_e = {'2013' : 100, '2014' : 200}
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = {'prod_a' : prod_a, 'prod_b' : prod_b, 'prod_c' : prod_c, 'prod_d' : prod_d, 'prod_e' : prod_e}

df_products = pd.DataFrame(data).transpose()
df_products.index.name = 'Products'
df_products = df_products.stack().reset_index()
df_products.columns = ['Product', 'Year', 'Value']

plt.figure(figsize = (10, 6))
sns.lineplot(data = df_products, x = 'Year', y = 'Value', hue = 'Product')
plt.title('Average Retail Product Price per Year')
```
